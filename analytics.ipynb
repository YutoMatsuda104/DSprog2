```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# データの読み込み
house_data = pd.read_csv('/Users/yutomatsuda/Lecture/DSProg2/data/horse_data.csv')
# NaNを含む列を削除
house_data = house_data.dropna(axis=1)

# 馬名をラベルエンコーディング
label_encoder = LabelEncoder()
house_data['馬名'] = label_encoder.fit_transform(house_data['馬名'])

# 目的変数と説明変数を分割
X = house_data.drop(['勝率平均'], axis=1)  # '勝率平均' 列を除外して説明変数を取得
y = house_data['勝率平均']  # '勝率平均' 列を目的変数として取得

# ランダムフォレストモデルのトレーニング
model = RandomForestRegressor()
model.fit(X, y)

# 各馬の予測値を得る
horse_names = house_data['馬名']  # 馬の名前を取得
predictions = model.predict(X)  # 各馬の予測値を得る

# 馬名と予測値を表示
for horse_name, prediction in zip(horse_names, predictions):
    print(f"馬名: {horse_name}, 予測勝率: {prediction}")
```

```python
from itertools import permutations

# 馬名と予測勝率をタプルのリストにする
horse_predictions = list(zip(horse_names, predictions))

# 予測勝率が高い順にソート
sorted_horse_predictions = sorted(horse_predictions, key=lambda x: x[1], reverse=True)

# 予測勝率の高い馬から3頭ずつ選んで順列を生成
top_horses = sorted_horse_predictions[:3]  # 予測勝率の高い馬から3頭選ぶ
permutations_3 = list(permutations(top_horses, 3))  # 3頭の順列を生成

# 三連単の買い目を表示
for combination in permutations_3:
    horse_1, horse_2, horse_3 = combination
    print(f"三連単: {horse_1[0]} - {horse_2[0]} - {horse_3[0]}")
```

```python
from itertools import permutations

# 馬名と予測勝率をタプルのリストにする
horse_predictions = list(zip(horse_names, predictions))

# 予測勝率が高い順にソート
sorted_horse_predictions = sorted(horse_predictions, key=lambda x: x[1], reverse=True)

# 予測勝率の高い馬から3頭ずつ選んで順列を生成
top_horses = sorted_horse_predictions[:100]  # 予測勝率の高い馬から100頭選ぶ
permutations_3 = list(permutations(top_horses, 3))  # 3頭の順列を生成

# 三連単の買い目を保存するリスト
trifecta_bets = []

# 三連単の買い目を生成
for combination in permutations_3:
    horse_1, horse_2, horse_3 = combination
    trifecta_bet = f"{horse_1[0]} - {horse_2[0]} - {horse_3[0]}"
    trifecta_bets.append(trifecta_bet)

# 三連単の買い目をCSVファイルに保存
import csv

with open('top_100_trifecta_bets.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["三連単"])
    for bet in trifecta_bets:
        writer.writerow([bet])

print("TOP100の三連単の買い目をCSVファイルに保存しました。")
```
