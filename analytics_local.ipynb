```python
import pandas as pd
from itertools import permutations

# CSVファイルのパス
file_path = '/Users/yutomatsuda/Lecture/DSProg2/horse_local.csv'

# CSVファイルを読み込む
df = pd.read_csv(file_path)

# 各馬の1着、2着、3着の予測数の合計を計算
df['total_pred'] = df['first_pred'] + df['second_pred'] + df['third_pred']

# 合計予測数で並べ替え
sorted_df = df.sort_values(by='total_pred', ascending=False)

# トップ1000の馬を取得
top_horses = sorted_df.head(1000)['house_name'].tolist()

# これらの馬の全ての順列を生成
permutations_of_horses = list(permutations(top_horses, 3))

# 最も人気のある組み合わせトップ10を選択
popular_combinations = permutations_of_horses[:1000]

# 結果を表示
for i, combo in enumerate(popular_combinations, 1):
    print(f"{i}. {combo}")
```

```python
# 結果をDataFrameに変換
results_df = pd.DataFrame(popular_combinations, columns=['1st Place', '2nd Place', '3rd Place'])

# 結果をCSVファイルとして保存
output_file_path = '/Users/yutomatsuda/Lecture/DSProg2/localdata_top100.csv'
results_df.to_csv(output_file_path, index=False)

print(f"結果は {output_file_path} に保存されました。")
```
